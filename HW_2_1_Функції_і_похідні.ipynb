1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = sp.symbols('x')

fx = 10*x - 5
df = sp.diff(fx, x)
print('Похідна дорівнює', df)
```

```python
x = np.array(range(0,11))
fx = 10*x - 5
df = np.linspace(10, 10, num=11)
plt.plot(x, fx, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label='f\'(x) = 10', color='red')
plt.legend()
plt.show()
```

Чим більше годин витрачається тим оцінка вища. Похідна показує що темп росту оцінки незмінний (хоч 1 годину вчиться хоч 10)

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
fx1 = sp.sin(x)
fx2 = sp.ln(x)

df1 = sp.diff(fx1, x)
df2 = sp.diff(fx2, x)
print(f'f(x) = sin(x), f\'(x) = {df1}')
print(f'f(x) = ln(x), f\'(x) = {df2}')

x_val = np.linspace(-2 * np.pi, 2 * np.pi, 500)
df1_val = np.cos(x_val)
fx1_val = np.sin(x_val)

plt.plot(x_val, fx1_val, label='f(x) = sin(x)', color='blue')
plt.plot(x_val, df1_val, label='f\'(x) = cos(x)', color='red')
plt.legend()
plt.show()
```

```python
x_val = np.linspace(1, 100, 500)

df2_val = np.log(x_val)
fx2_val = 1/x_val

plt.plot(x_val, fx2_val, label='f(x) = 1/x', color='blue')
plt.plot(x_val, df2_val, label='f\'(x) = ln(x)', color='red')
plt.legend()
plt.show()
```
